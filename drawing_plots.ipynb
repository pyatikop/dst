```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
def histograms(data, plot_name):
    labels = ['1', '2', '3']
    colors = ['blue', 'red', 'gold', 'green', 'violet']

    x = np.arange(len(labels))
    width = 0.2

    fig, ax = plt.subplots()
    rects = []

    for i, (model, values) in enumerate(data.items()):
        rect = ax.bar(x + i * width, values, width, label=model, color=colors[i])
        rects.append(rect)

    ax.set_ylabel('R2 score')
    ax.set_xlabel('час')
    ax.set_title(plot_name)
    ax.set_xticks(x + width * 1.5)
    ax.set_xticklabels(labels)
    ax.axhline(y=0, color='k')
    ax.legend()
    ax.set_ylim(-0.6 ,1)

    fig.tight_layout()
    plt.show()
```

```python
plot_name = 'ACE на ACE, автокорреляционная функция'
data = {
    'LinReg': [-0.507, -0.647, -0.212],
    'Ridge-Reg, L=10^6': [0.717, 0.639, 0.574],
    'Ridge-Reg, L=5*10^6': [0.861, 0.762, 0.642],
    'Ridge-Reg, L=10^7': [0.865, 0.759, 0.639],
    'CatBoost': [0.950, 0.875, 0.795]
}
```

```python
histograms(data, plot_name)
```
